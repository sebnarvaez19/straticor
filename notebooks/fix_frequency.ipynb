```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import straticor as scr

from scipy.interpolate import interp1d

plt.style.use("https://raw.githubusercontent.com/srnarvaez/my-mpl-themes/main/basic.mplstyle")

%matplotlib widget
```

```python
column = scr.load_column("../data/afloramiento/quebrada_santa_lucia/quebrada_santa_lucia.xlsx")
fixed = scr.load_column("../data/afloramiento/quebrada_santa_lucia/quebrada_santa_lucia.xlsx", fix=True)

fig, ax = plt.subplots(figsize=(5, 8))

scr.plot_column(column, ax, 0.5, 0.5, lw=0, legend=False)
ax.step(column["environment"] / 7 + 1, column["bottom"], where="post", lw=1.0, color="k")

scr.plot_column(fixed, ax, 2.5, 0.5, lw=0, legend=False)
ax.plot(fixed["environment"] / 7 + 3, fixed["bottom"], lw=1.0, color="r")

ax.set(xlim=(0, 4.2), title="Quebrada Santa Lucía", ylabel="Height [ft]", xticks=[])

ax.text(0.75, 110, "Original", ha="center", va="center")
ax.text(2.75, 110, "Processed", ha="center", va="center")

scr.add_legend(column, ax, loc="center left", bbox_to_anchor=(1.2, 0.5))

fig.savefig("../images/fixed_quebrada_santa_lucia.png", dpi=300, bbox_inches="tight")

plt.show()
```
